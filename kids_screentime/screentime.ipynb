```python

labels=['1-10','11-15','16-18']
new_df['Age_groups']=pd.cut(new_df['Age'],bins=[0,10,15,18],labels=labels)
```

```python
#removing age column
new_df.drop(columns=['Age'],inplace=True)
```

```python
new_df.sample(2)
```

```python
new_df.Exceeded_Recommended_Limit.value_counts()
```

```python
for i, pre in enumerate(new_df.drop(columns=["Exceeded_Recommended_Limit", "Educational_to_Recreational_Ratio", "Avg_Daily_Screen_Time_hr"])):
    plt.figure(i)
    s = sns.countplot(data=new_df, hue="Exceeded_Recommended_Limit", x=pre)
    plt.xticks(rotation=90)
    handles, labels = s.get_legend_handles_labels()
    new_labels = ["Within Limit" if lbl == "0" else "Time Limit Exceeded" for lbl in labels]
    plt.legend(handles=handles, labels=new_labels, title="Exceeded Limit", loc="center left", bbox_to_anchor=(1, 0.5))
    plt.tight_layout()
    for cont in s.containers:
        s.bar_label(cont)
```

```python
new_df_dummies=pd.get_dummies(new_df)
exclude = ['Avg_Daily_Screen_Time_hr', 'Educational_to_Recreational_Ratio']

# Convert all columns except the excluded ones to int
new_df_dummies[new_df_dummies.columns.difference(exclude)] = (
    new_df_dummies[new_df_dummies.columns.difference(exclude)].astype(int)
)
new_df_dummies.head(5)
```

```python
plt.figure(figsize=(20,8))
new_df_dummies.corr()['Avg_Daily_Screen_Time_hr'].sort_values(ascending=False).plot(kind='bar')
```

```python
sns.boxplot(x='Gender', y='Avg_Daily_Screen_Time_hr', data=new_df, hue='Gender', palette='Set2')

plt.title('Screen Time Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Avg Daily Screen Time (hrs)')
plt.tight_layout()
plt.show()
```
